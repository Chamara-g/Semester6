```python
import numpy as np

fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(hspace=.5) 
axes = plt.subplot(121)

x = np.linspace(0, 10, 100)
y = np.cos(x)

axes.plot(x,y,linewidth=4.0, ls='--', color='r', marker="o")

axes.legend("cos(x)",loc="upper right")
axes.set_title("Title for (5,1) subplot")
axes.set_xlabel('xlabel')
axes.set_ylabel('ylabel')

axes.set_autoscaley_on(False)
axes.set_ylim([-2,2])
axes.xaxis.set(ticks=range(1,10))
fig.text(0.5, 0.04, 'common x label', ha='center', fontsize=15)
fig.text(0.04, 0.5, 'common y label', va='center', rotation='vertical', fontsize=15)
```

```python
import numpy as np

fig = plt.figure(figsize=(15,10))
fig.subplots_adjust(hspace=.5) 
axes = plt.subplot(121)
axes1 = fig.add_subplot(122)

x = np.linspace(0, 10, 100)
y = np.cos(x)
y1 = np.tan(x)

axes.plot(x,y,linewidth=4.0, ls='--', color='r', marker="o")

axes.legend("cos(x)",loc="upper right")
axes.set_title("Title for (5,1) subplot")
axes.set_xlabel('xlabel')
axes.set_ylabel('ylabel')

axes.set_autoscaley_on(False)
axes.set_ylim([-2,2])
axes.xaxis.set(ticks=range(1,10))

axes1.plot(x,y1,linewidth=4.0, ls='--', color='r', marker="o")

axes1.legend("sin(x)",loc="upper right")
axes1.set_title("Title for (5,1) subplot")
axes1.set_xlabel('xlabel')
axes1.set_ylabel('ylabel')

axes1.set_autoscaley_on(True)

fig.text(0.5, 0.04, 'common x label', ha='center', fontsize=15)
fig.text(0.04, 0.5, 'common y label', va='center', rotation='vertical', fontsize=15)

plt.savefig('cosineAndtan.png', dpi=400, bbox_inches='tight')
```

```python
import numpy as np
import pylab
import random
import matplotlib
import matplotlib.pyplot as plt

class RandomWalk:
    def __init__(self,steps):
        self.steps = steps

    def randomPath(self,X_start,Y_start):
        currentX = X_start;
        currentY = Y_start;

        x_coordinates = list()
        y_coordinates = list()

        for i in range(self.steps):
            isGoing = np.random.randint(2) #0 wait, 1 going
            side = np.random.randint(4) # 0 front 1 back 2 up 3 down

            if side == 0:
                if currentX != self.steps: 
                    currentX += isGoing #going front

            elif side == 1:
                if currentX != 0:
                    currentX -= isGoing #going back

            elif side == 2:
                if currentY != 0:
                    currentY -= isGoing #going up

            elif side == 3:
                if currentY != self.steps:
                    currentY += isGoing #going down

            x_coordinates.append(currentX)
            y_coordinates.append(currentY)
            
            #print('step:',i+1,'side:',side,'isGoing:',isGoing,'X:',currentX,'Y:',currentY)        
            
        return x_coordinates,y_coordinates
            
randomWalk = RandomWalk(500)
Pos_v1_x,Pos_v1_y = randomWalk.randomPath(300,0)

fig = plt.figure(figsize=(18,4))
fig.subplots_adjust(hspace=0.05)
axes1= fig.add_subplot(131)
axes1.plot(Pos_v1_x)
axes1.set_title("Normal plot")

axes2= fig.add_subplot(132)
axes2.hist(Pos_v1_x , bins=20, color='r')
axes2.set_title("Histogram plot")

axes3= fig.add_subplot(133)
axes3.scatter(Pos_v1_x,Pos_v1_y , color='green', alpha=0.3, edgecolors='grey')
axes3.set_title("scatter plot")

line_labels = ["normal", "histogram", "scatter"]


fig.legend( labels=line_labels,   # The labels for each line
           loc="center right",   # Position of legend
           borderaxespad=0.1,    # Small spacing around legend box
           title="All subplots legend "  # Title for the legend
           )

plt.show()


```
